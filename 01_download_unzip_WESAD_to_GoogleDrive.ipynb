# 01 — Download and Unzip WESAD Dataset to Google Drive

This notebook only downloads the WESAD dataset from Kaggle into Google Drive and unzips it. It does **not** preprocess or model the data.

The Kaggle API credential file, `kaggle.json`, is required to access and download the dataset.

```
# STEP 3. Upload kaggle.json
# Choose the kaggle.json file downloaded from Kaggle Legacy API Credentials.
from google.colab import files
uploaded = files.upload()

print('Uploaded files:', list(uploaded.keys()))
```

```
# STEP 4. Configure Kaggle API credentials
import os, shutil

os.makedirs('/root/.kaggle', exist_ok=True)

if 'kaggle.json' not in os.listdir('/content'):
    raise FileNotFoundError('kaggle.json was not found in /content. Please upload kaggle.json in the previous cell.')

shutil.copy('/content/kaggle.json', '/root/.kaggle/kaggle.json')
os.chmod('/root/.kaggle/kaggle.json', 0o600)

print('Kaggle API configured successfully.')
```

```
# STEP 5. Test Kaggle connection
!kaggle datasets list -s WESAD
```

```
# STEP 6. Create project folder in Google Drive
import os

DRIVE_ROOT = '/content/drive/MyDrive/Apple_Hidden_Wellness_AI'
RAW_DATA_DIR = os.path.join(DRIVE_ROOT, 'data', 'raw', 'WESAD')

os.makedirs(RAW_DATA_DIR, exist_ok=True)

print('Dataset will be saved to:')
print(RAW_DATA_DIR)
```

```
# STEP 7. Download WESAD dataset from Kaggle
# Selected Kaggle dataset slug from search results.
# Dataset identifier used for this analysis; update only if a different WESAD source is required.

DATASET_SLUG = 'orvile/wesad-wearable-stress-affect-detection-dataset'

!kaggle datasets download -d {DATASET_SLUG} -p "{RAW_DATA_DIR}"

print('Download complete. Files in folder:')
!ls -lh "{RAW_DATA_DIR}
```

```
# STEP 8. Unzip downloaded dataset
import glob, zipfile, os

zip_files = glob.glob(os.path.join(RAW_DATA_DIR, '*.zip'))
print('ZIP files found:', zip_files)

if len(zip_files) == 0:
    raise FileNotFoundError('No ZIP file found. Please check whether the Kaggle download completed successfully.')

for zip_path in zip_files:
    print('Extracting:', zip_path)
    with zipfile.ZipFile(zip_path, 'r') as zip_ref:
        zip_ref.extractall(RAW_DATA_DIR)

print('Unzip complete.')
```

```
# STEP 9. Inspect dataset structure
import os

for path, dirs, files in os.walk(RAW_DATA_DIR):
    level = path.replace(RAW_DATA_DIR, '').count(os.sep)
    indent = '  ' * level
    print(f'{indent}{os.path.basename(path)}/')
    subindent = '  ' * (level + 1)
    for f in files[:10]:
        print(f'{subindent}{f}')
    if level >= 3:
        dirs[:] = []
```

```
# STEP 10. Search for key WESAD files
import glob, os

pkl_files = glob.glob(os.path.join(RAW_DATA_DIR, '**', '*.pkl'), recursive=True)
respiban_files = glob.glob(os.path.join(RAW_DATA_DIR, '**', '*respiban*.txt'), recursive=True)
quest_files = glob.glob(os.path.join(RAW_DATA_DIR, '**', '*quest*.csv'), recursive=True)
e4_files = glob.glob(os.path.join(RAW_DATA_DIR, '**', '*E4*'), recursive=True)

print('PKL files:', len(pkl_files))
print('Respiban files:', len(respiban_files))
print('Quest files:', len(quest_files))
print('E4-related files/folders:', len(e4_files))

print('Example files:')
for f in (pkl_files[:5] + respiban_files[:5] + quest_files[:5]):
    print(f)
```
